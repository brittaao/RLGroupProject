```python
import matplotlib.pyplot as plt 
import pandas as pd

fig, ax = plt.subplots()

for i in range(1,9):
    data = pd.read_csv(f'train_results/model{i}.csv', header=None)

    data = data.transpose()
    #ax.plot(range(1,1000,25), data)
    ax.plot(range(1, 1000, 25), data, label=f'Model {i}')
plt.title('Cart Pole Training Results')
plt.xlabel('Episodes')
plt.ylabel('Average Return')
ax.legend()
#plt.figure(figsize=(300,50))
plt.savefig('CartPoleTraining.png')
plt.show()
```
